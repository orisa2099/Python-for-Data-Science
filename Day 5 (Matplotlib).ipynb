# Data Visualization

## Matplot Lib


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,20)    #generating numbers that are linearly spaced numbers
x
```

```python
y = np.cos(x)
y
```

### functional plotting

```python
#function plot; this invloves plotting on the default plot(the box)
#oop plot invloves modifying the plot area and characteristics to your taste
plt.plot(x,y,'g:')
plt.title('Title goes here', fontweight='bold', fontsize=16)
plt.ylabel('Y Label title goes here')
plt.xlabel('X label title goes here')
plt.show()     #this displays the plot not as an output but as a printed object
```

```python
## subplot 
plt.subplot(1,2,1) # first argument:no of rows, second argument: no of columns, third argument is the hierachy of plot,1 is the first plot
plt.plot(x,y,'r--')
plt.subplot(1,2,2)
plt.plot(x,np.sin(x), 'g*-');   # the ; is another shortcut to plt.show
plt.tight_layout(); # to spread out the graph from each other
```

### Objected Oriented plotting

```python
# create an empty canvas
fig = plt.figure()

#add axes to figure created
axes = fig.add_axes([0.1,0.1,0.8,0.8])   #left, bottom(to the ground), width, height

#adding plot on the created canvas
axes.plot(x,y,'b:')
axes.set_title('Title name', fontsize=16)
axes.set_xlabel('x labelling')
axes.set_ylabel('y labelling')
```

```python
fig = plt.figure()

#add axes to figure created
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])  #main axes plot
axes2 = fig.add_axes([0.83,0.3,0.4,0.5])

#larger canas axes 1
axes1.plot(x,y,'cyan')
axes1.set_title('Title1 name', fontsize=16)
axes1.set_xlabel('x1 labelling')
axes1.set_ylabel('y1 labelling')

#smaller canvas
axes2.plot(x,np.sin(x),'r:')
axes2.set_title('Title2 name', fontsize=14)
axes2.set_xlabel('x2 labelling')
axes2.set_ylabel('y2 labelling')
```

### Legends

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y, label='Cosine')
ax.plot(x,np.sin(x), label='sine')
ax.legend(loc=4)   #loc=location of legend
```

### Setting colours, line width and linetypes

```python
fig, axes = plt.subplots()
axes.plot(x,y, color='blue', lw=3 )   #lw is line width, ls is line styel(-,_,-.,:,*-)
axes.plot(x,np.sin(x), color='yellow', alpha=0.8)   # alpha is to make color transparent 
axes.plot(x,x*y, color='#FF8C00', ls = "dashdot")
```

### Markers

```python
fig, axes = plt.subplots()
axes.plot(x,y, color='blue', lw=3, ls = "dashdot", marker="o", markersize=10,
         markerfacecolor='black', markeredgecolor='green')  # market is to pick the data point in the chart
```

### control over axes

```python
fig, axes = plt.subplots()
axes.plot(x,y, color='blue',)
axes.set_xlim([1.2,8])   #setting the range for the x-axis
axes.set_ylim([-1,1.5])
```

```python
software =('python','Excel','R','SPSS','SQL','JS','Hadoop')
Rating = [15,8,12,10,6,11,5]
color =['r','g','b','y','coral','k','lightblue']
plt.bar(software,Rating, color = color, alpha=0.5)
```

### Pie

```python
software =('python','Excel','R','SPSS')
sizes = [215,130,245,210]
color =['r','g','b','y']
explode=(0,0.1,0,0)
plt.pie(sizes,labels=software,colors=color,autopct='%.1f%%',explode=explode, startangle=90)
```

## Statistics plot

```python
from statsmodels.graphics.gofplots import qqplot
from scipy.stats import shapiro
import numpy as np
```

```python
data = 5*np.random.rand(100)+50
data
```

```python
plt.hist(data)
```

```python
qqplot(data, line='s')
```

```python
stat,pvalue=shapiro(data)
```

```python
print('statistics=%.3f, p_value = %.3f'%(stat,pvalue))
```

```python
alpha = 0.05
if pvalue > alpha:
    print('sample look guassian i.e normal')
else:
    print('sample is not normal')
```
