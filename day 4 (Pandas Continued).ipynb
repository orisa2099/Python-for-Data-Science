# MERGING JOINING AND CONCATENATING

```python
import pandas as pd
```

```python
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']
                   },index=[0,1,2,3])


df1
```

```python
df2 =pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']
                   },index=[4,5,6,7])
df2
```

```python
df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']
                   },index=[8,9,10,11])
df3
```

## Concatenanting

```python
pd.concat([df1,df2,df3])   #row-wise contactenation same rows and differing columns
```

```python
pd.concat([df1,df2,df3], axis=1)  #column-wise contactenation: same column and differing rows
```

## Merging

```python
left = pd.DataFrame({'Key':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']
                   },index=[0,1,2,3])

left
```

```python
right =pd.DataFrame({'Key':['A0','A1','A2','A3'],
                   'E':['B4','B5','B6','B7'],
                   'F':['C4','C5','C6','C7'],
                   'G':['D4','D5','D6','D7']
                   },index=[0,1,2,3])
right
```

```python
pd.merge(left,right,on='Key',how='inner')
```

## Joining 

to join two data set of same index though different column and vice-versa

```python
df1 =pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3']},
                    index=['K0','K1','K2','K3'])
df1
```

```python
df2 =pd.DataFrame({'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']},
                    index=['K0','K1','K3','K4'])
df2
```

```python
df1.join(df2)  #automatically removes eoither column or index that are not same in both dataset
```
