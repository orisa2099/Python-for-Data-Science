#### Types of Tokenization 

sentence tokenization: paragrapgh into sentence


word tokenization: sentence to word

```python
from nltk.tokenize import sent_tokenize
```

```python
nltk.download('punkt')
```

```python
text = """Hello Mr. Yemi, How are you today. I know you are a good man and you will enter heaven.
I know you will reject some porject today and tommorow and God will continue to help you"""
```

```python
tok_sent = sent_tokenize(text)
```

```python
type(tok_sent)
```

```python
from nltk.tokenize import word_tokenize
```

```python
tok_word = word_tokenize(text)
print(tok_word)
```

### Word frequency Distribution

```python
from nltk.probability import FreqDist
```

```python
fdist = FreqDist(tok_word)
print(fdist)
```

```python
fdist.most_common(5)
```

### Frequency Distribution plot

```python
import matplotlib.pyplot as plt
```

```python
fdist.plot(41, cumulative = False)
```

### Stopwords

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
```

```python
stopword = set(stopwords.words('english'))
print(stopword)
```

### Removing stopword from a text



```python
filtered_sent=[]
for i in tok_word:
    if i not in stopword:
        filtered_sent.append(i)
    else:
        pass

print('Tokenize word:', tok_word)
print('Filtered Sent:', filtered_sent)
```

## Lexicon Normalization

types of lexicon normalization
1. lemar; this reduces word to their base word which linguistically correct
2. stemming is a process of linguistic normalization, reducing word to their word root

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
stemmedword = []
for w in filtered_sent:
    stemmedword.append(ps.stem(w))
    

print('Filtered Sent:', filtered_sent)    
print('stemmedword:', stemmedword)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
lem = WordNetLemmatizer()
```

```python
word = "flying"

lem.lemmatize(word,'v')
```

```python
lem_word =[]
for w in filtered_sent:
    lem_word.append(lem.lemmatize(w,'v'))
    
print('Stemmed sentence', stemmedword)
print('lem sentence', lem_word)
```

```python
from wordcloud import WordCloud, STOPWORDS
```

```python
comment=''
stopwords = set(STOPWORDS)
```

```python
df.CONTENT
```

```python
for value in df.CONTENT:
    #converting each value to a string
    value = str(value)
    
    #split the value
    tokens = value.split()
    
    #covert to lower case
    for i in range(len(tokens)):
        tokens[i] = tokens[i].lower()
        
    comment += " ".join(tokens)+" "
    
```

```python
wordcloud = WordCloud(width =800, height=800, background_color='navy', colormap='rainbow',
                      random_state=1,stopwords=stopwords, min_font_size=10).generate(comment)
```

```python
import matplotlib.pyplot as plt
```

```python
# plotting the word cloud image

plt.figure(figsize=(8,8), facecolor=None)
plt.imshow(wordcloud)
plt.axis('off')
plt.tight_layout(pad=0)
plt.show()
```
