### Joint Plot

```python
sns.jointplot(x='total_bill',y='tip', data =df, kind='reg')   
# kind is to change the relationship curve you want to use e,g 
#reg(regression)
#kde(focus on where the union is heavy)
#hex(hexagon)
#resid(devides the datapoint into two equa;l halves)
```

### Pairplot

takes all the numerical in the dataframe and plot thier interaction

```python
sns.pairplot(df, hue='sex', palette='coolwarm')
#hue helps separate the plot based on categorical variable given
```

### Rug plot

this are base plot for the kde, they plot the upper part of the Kde curve

```python
sns.kdeplot(df['total_bill'])
sns.rugplot(df['total_bill'])
```
